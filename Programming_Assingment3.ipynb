### Ques(1): Write a Python Program to Check if a Number is Positive, Negative or Zero?
### Ans-


```python
number=float(input("Enter the number:"))
if number>0:
    print("Entered Number",number,"is positive.")
elif number<0:
    print("Entered Number",number,"is negative.")
else:
    print("Entered Number",number,"is zero.")
```

### Ques(2): Write a Python Program to Check if a Number is Odd or Even?
### Ans-

```python
number=int(input("Enter the number:"))
if number%2==0:
    print("Entered Number",number,"is Even.")
else:
    print("Entered Number",number,"is Odd.")
```

### Ques(3): Write a Python Program to Check Leap Year?
### Ans-

```python
year=int(input("Enter the year."))
if (year%400==0) and (year%100==0):
    print("Entered year",year,"is a Leap Year.")
elif (year%4==0) and (year%100!=0):
    print("Entered year",year,"is a Leap Year.")
else:
    print("Entered year",year,"is not a Leap Year.")
```

### Ques(4): Write a Python Program to Check Prime Number?
### Ans-

```python
num=int(input("Enter the Number: "))
count=0 #flag=False
if num>1:
    for i in range(2,num):
        if (num%i==0):
            count=1 #flag=True
            break
if count==1:#flag=True
    print("Entered number",num,"is not a Prime Number.")
else:
    print("Entered number",num,"is a Prime Number.")
```
